# Missing Value

## Data

```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame(
    {
        'x1':[4,6,7,np.nan,2,1],
        'x2':[1,4,6,8,np.nan,1],
        'x3':[11,12,13,16,18,20],
        'x4':['A','A',np.nan,'C','C','D'],
        'x5':['X','X','Y',np.nan,'X','Y'],
        'x6':['M','M',np.nan,np.nan,'N','N'],
    }
)
```

```python
df
```

## Mean & Median (Numeric)

```python
from sklearn.impute import SimpleImputer
```

```python
simple_impute = SimpleImputer(strategy='mean')
df[['x1', 'x2']] = simple_impute.fit_transform(df[['x1','x2']])
df
```

## Mode (categoric)

```python
simple_impute = SimpleImputer(strategy='most_frequent')
df[['x4','x5']] = simple_impute.fit_transform(df[['x4','x5']])
df
```

## Constant (Numeric & Categoric)

```python
df1 = pd.DataFrame(
    {
        'x1':[4,6,7,np.nan,2,1],
        'x2':[1,4,6,8,np.nan,1],
        'x3':[11,12,13,16,18,20],
        'x4':['A','A',np.nan,'C','C','D'],
        'x5':['X','X','Y',np.nan,'X','Y'],
        'x6':['M','M',np.nan,np.nan,'N','N'],
    }
)
```

```python
simple_impute = SimpleImputer(strategy='constant', fill_value=5)
df1[['x1','x2']] = simple_impute.fit_transform(df1[['x1','x2']])
df1
```

```python
simple_impute = SimpleImputer(strategy='constant', fill_value='P')
df1[['x4','x5','x6']] = simple_impute.fit_transform(df1[['x4','x5','x6']])
df1
```

## Iterative Imputer

1. For numerical value
2. Predictive approach
3. Linear model

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
```

```python
df2 = pd.DataFrame(
    {
        'x1':[4,6,7,np.nan,2,1],
        'x2':[1,4,6,8,np.nan,1],
        'x3':[11,12,13,16,18,20],
        'x4':['A','A',np.nan,'C','C','D'],
        'x5':['X','X','Y',np.nan,'X','Y'],
        'x6':['M','M',np.nan,np.nan,'N','N'],
    }
)
```

```python
iter_impute = IterativeImputer()
df2[['x1', 'x2', 'x3']] = iter_impute.fit_transform(df2[['x1', 'x2', 'x3']])
df2
```
