# Polynomial Features

## Library

```python
# data
import pandas as pd
import numpy as np

# visualization
import seaborn as sns
import matplotlib.pyplot as plt

# preprocessing
from sklearn.preprocessing import PolynomialFeatures

# data splitting
from sklearn.model_selection import train_test_split

# modeling
from sklearn.linear_model import LogisticRegression


# validation
from sklearn.metrics import accuracy_score
```

## Data

```python
wine = pd.read_csv('./datasets/winequality-white.csv', delimiter=';')
wine['label'] = np.where(wine['quality'] > 6, 1, 0)
wine
```

## Data Splitting

```python
X = wine[['alcohol', 'density']]
y = wine['label']
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    stratify=y,
    random_state=2020
)
```

## Model without Polynomial Feature

```python
logreg = LogisticRegression()
logreg.fit(X_train, y_train)
```

```python
y_pred = logreg.predict(X_test)
accuracy_score(y_test, y_pred)
```

## Model with Polynomial Features

```python
polynom = PolynomialFeatures(degree=3, include_bias=False)
X_train_poly = polynom.fit_transform(X_train)
X_test_poly = polynom.transform(X_test)
```

```python
pd.DataFrame(X_train_poly, columns=polynom.get_feature_names())
```

```python
logreg = LogisticRegression()
logreg.fit(X_train_poly, y_train)
```

```python
y_pred_poly = logreg.predict(X_test_poly)
accuracy_score(y_test, y_pred_poly)
```
