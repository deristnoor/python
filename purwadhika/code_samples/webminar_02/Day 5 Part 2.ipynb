# **Recommendation System**

## **Collaborative filtering**

collaborative.png

- **Rekomendasi produk diberikan berdasarkan past interactions antara users dan items** â€”> Interaksi tersebut disimpan di 'user-item interactions matrix'.
- Jadi, ide utama dari collaborative filtering ini adalah user-item interactions yang sudah terjadi dapat digunakan untuk mendeteksi users atau items yang mirip lalu dibuatlah prediksi rekomendasi berdasarkan hal tersebut.
- Collaborative filtering terbagi menjadi 2 kategori, yaitu *memory based* dan *model based*.
  - **Memory based** --> user-user dan item-item (metode dengan low bias, high variance)
  - **Model based** --> matrix factorization (model dengan high bias, lower variance)
<br><br>
- Memory based bekerja secara langsung berdasarkan interaksi yang sudah terekam, dengan asumsi tidak memakai bantuan model, tetapi hanya berdasarkan pencarian nearest neighbours saja.
- Sebagai contoh, rekomendasi dibuat dengan mencari users terdekat dari suatu user yang memiliki interest yang mirip, lalu dibuatlah rekomendasi terhadap items yang paling popular di antara neighbours atau user terdekat tersebut.
- Sedangkan pendekatan model based menggunakan asumsi bahwa model akan menjelaskan user-item interactions dan mencoba menemukan pola dari interaction tersebut untuk dapat membuat prediksi rekomendasi baru.
<br><br>
- Keunggulan utama dari collaborative filtering adalah tidak dibutuhkannya informasi spesifik terkait users atau items. Dengan kata lain, semakin banyak users yang berinteraksi dengan intems, semakin akurat pula rekomendasi baru yang dihasilkan.
- Namun, karena collaborative filtering mengandalkan interaksi yang sudah terjadi untuk dapat membuat rekomendasi, metode ini tidak bisa mengatasi 'cold start problem' atau kondisi di mana ada user atau item baru yang belum memiliki data interaksi yang telah terjadi, sehingga rekomendasi tidak bisa dibuat. Dengan kata lain, kita tidak bisa merekomendasikan item baru ke user manapun, dan user baru pun tidak bisa mendapatkan rekomendasi apa-apa.
- Untuk mengatasi cold start problem, ada beberapa strategi yang dapat digunakan, yaitu:
    - Merekomendasikan random items kepada users baru atau items baru kepada random users (random strategy).
    - Merekomendasikan popular items kepada users baru atau items baru kepada users yang paling aktif (maximum expectation strategy).
    - Merekomendasikan satu set yang berisi macam-macam items kepada users baru atau sebaliknya (exploratory strategy).
    - Menggunakan metode non-collaborative untuk user dan item baru. Bisa dengan menggunakan content based method contohnya.


## **Content based**

content.png

- Tidak seperti metode collaborative yang hanya mengandalkan user-item interactions, metode content based menggunakan informasi tambahan terkait users atau items.
- Sebagai contoh, jika kita membuat movies recommender system, informasi tambahannya adalah: 
  - Untuk user bisa dilihat dari umur, jenis kelamin, pekerjaan atau informasi personal lainnya. 
  - Untuk item bisa dilihat dari pemeran utama, durasi, genre dan karakteristik lainnya.
<br><br>
- Jadi, ide utama dari metode content-based adalah mencoba membangun model rekomendasi berdasarkan features yang ada, yang secara implisit menjelaskan interaksi antara users dan items.
- Namun, interaksi yang tercipta untuk dipelajari oleh model berasal dari semua users, sehingga users yang sebenarnya memiliki karakteristik yang sama pun bisa saja memiliki preferensi yang berbeda. Dengan kata lain, metode ini sebenarnya kurang 'personal' (menghasilkan bias yang lebih tinggi).
<br><br>
- Metode content-based tidak memiliki permasalahan cold start problem. Users atau items baru dapat dijelaskan berdasarkan karakterisitknya, sehingga rekomendasi yang sesuai dapat diberikan terhadap users atau items baru tersebut.
- Seperti yang sudah disebutkan di poin sebelumnya, metode content based secara teoritis menghasilkan bias yang paling tinggi, tetapi variance yang paling rendah dibandingkan dengan metode collaborative. Hal ini dikarenakan metode ini menggunakan asumsi model, di mana informasi yang ada merepresentasikan users atau items secara spesifik, sehingga cenderung menghasilkan bias yang lebih tinggi, tetapi variance yang lebih rendah.
