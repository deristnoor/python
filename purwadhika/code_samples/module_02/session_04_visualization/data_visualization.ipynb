# 1. Histogram
### Creating Histogram Using Matplotlib

```python
import matplotlib.pyplot as plt
import seaborn as sns

tips = sns.load_dataset('tips')
tips.head()
```

```python
plt.hist(tips['total_bill'])
plt.title('Total Bill Histogram', size=15)
plt.xlabel('Total Bill')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.style.use('seaborn')
plt.figure(figsize=(8,6))
plt.hist(tips['total_bill'], bins=10, color='green')
plt.title('Total Bill Histogram', size=15)
plt.xlabel('Total Bill', size=10)
plt.ylabel('Frequency', size=10)
plt.grid(True)
plt.legend(['Total Bill from Customer'], loc=0)
plt.savefig('total_bill_histogram.png')
plt.show()
```

### Creating Histogram Using Seaborn

```python
sns.distplot(tips['tip'])
plt.title('Tip Histogram', size=15)
plt.show()
```

```python
sns.displot(tips['tip'])
plt.title('Tip Histogram', size=15)
plt.show()
```

```python
sns.histplot(tips['tip'])
plt.title('Tip Histogram', size=15)
plt.show()
```

### Creating Histogram Using Pandas

```python
tips['tip'].hist(color='green', alpha=0.4)
plt.title('Tip Histogram', size=15)
plt.show()
```

# 3. Lineplot

### Create Lineplot using Matplotlib

```python
import matplotlib.pyplot as plt
import seaborn as sns

flights = sns.load_dataset('flights')
flights.head()
```

```python
x = flights[flights['month'] == 'Jan']['year']
y = flights[flights['month'] == 'Jan']['passengers']

plt.figure(figsize=(8,6))
plt.plot(x, y, 'blue')
plt.title('Number of Passengers Trends in January', size=15)
plt.xlabel('Year')
plt.ylabel('Number of Passengers')
plt.show()
```

```python
x1 = flights[flights['month'] == 'Jan']['year']
y1 = flights[flights['month'] == 'Jan']['passengers']
x2 = flights[flights['month'] == 'Dec']['year']
y2 = flights[flights['month'] == 'Dec']['passengers']

plt.style.use('seaborn')
plt.figure(figsize=(8,6))
plt.plot(x1, y1, 'blue', 
         x2, y2, 'red')
plt.title('Number of Passengers Trends in January & December', size=15)
plt.xlabel('Year')
plt.ylabel('Number of Passengers')
plt.legend(['Januar Passengers', 'December Passengers'])
plt.show()
```

### Create Lineplot using Seaborn

```python
sns.lineplot(data=flights, x='year', y='passengers')
plt.title('Number of Passengers Trends', size=15)
plt.show()
```

```python
sns.lineplot(data=flights, x='year', y='passengers', hue='month')
plt.title('Number of Passengers Trends by Months', size=15)
plt.show()
```

### Create Lineplot using Pandas

```python
df_jan = flights[flights['month'] == 'Jan']
df_jan.plot.line(x='year', y='passengers')
plt.title('Number of Passengers Trends in January', size=15)
plt.show()
```

# 4. Scatter Plot

### Create Scatterplot using Matplotlib

```python
import matplotlib.pyplot as plt
import seaborn as sns

tips = sns.load_dataset('tips')
tips.head()
```

```python
x = tips['total_bill']
y = tips['tip']

plt.figure(figsize=(8,6))
plt.scatter(x, y)
plt.title('Scatterplot: Total Bill vs Tip', size=15)
plt.xlabel('Total_bill')
plt.ylabel('Tip')
plt.show()
```

```python
x = tips['size']
y = tips['tip']

plt.style.use('seaborn')
plt.figure(figsize=(8,6))
plt.scatter(x, y)
plt.title('Scatterplot: Meal Size vs Tip', size=15)
plt.xlabel('Meal Size')
plt.ylabel('Tip')
plt.show()
```

### Create Scatterplot using Seaborn

```python
sns.scatterplot(data=tips, x='total_bill', y='tip')
plt.title('Scatterplot: Total Bill vs Tip', size=15)
plt.show()
```

```python
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='day')
plt.title('Scatterplot: Total Bill vs Tip', size=15)
plt.show()
```

```python
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='day', style='time')
plt.title('Scatterplot: Total Bill vs Tip', size=15)
plt.show()
```

### Create Scatterplot using Pandas

```python
tips.plot.scatter(x='total_bill', y='tip')
plt.title('Scatterplot: Total Bill vs Tip', size=15)
plt.show()
```

# 5. Bar Plot

### Create Barplot using Matplotlib

```python
import matplotlib.pyplot as plt
import seaborn as sns

tips = sns.load_dataset('tips')
tips.head()
```

```python
tips_by_day = tips[['tip', 'day']].groupby(['day']).mean()
tips_by_day
```

```python
x = tips_by_day.index.tolist()
y = tips_by_day['tip']

plt.figure(figsize=(8,6))
plt.bar(x, y, color=['green', 'blue', 'red', 'orange'])
plt.title('Barplot: Average Tip Each Day', size=15)
plt.xlabel('Day')
plt.ylabel('Tip')
plt.show()
```

```python
x = tips_by_day.index.tolist()
y = tips_by_day['tip']

plt.figure(figsize=(8,6))
plt.barh(x, y, color=['green', 'blue', 'red', 'orange'])
plt.title('Barplot: Average Tip Each Day', size=15)
plt.xlabel('Tip')
plt.ylabel('Day')
plt.show()
```

```python
x = tips_by_day.index.tolist()
y = tips_by_day['tip']

plt.style.use('seaborn')
plt.figure(figsize=(8,6))
plt.bar(x, y, color=['green', 'blue', 'red', 'orange'])
plt.title('Barplot: Average Tip Each Day', size=15)
plt.xlabel('Day')
plt.ylabel('Tip')
plt.show()
```

### Create Barplot using Seaborn

```python
sns.barplot(data=tips, x='day', y='tip')
plt.title('Barplot: Average Tip Each Day', size=15)
plt.show()
```

```python
sns.barplot(data=tips, x='day', y='tip', hue='sex')
plt.title('Barplot: Average Tip Each Day by Gender', size=15)
plt.legend(loc=0)
plt.show()
```

```python
sns.barplot(data=tips, y='day', x='tip', hue='sex')
plt.title('Barplot: Average Tip Each Day by Gender', size=15)
plt.legend(loc=0)
plt.show()
```

```python
import numpy as np

sns.barplot(data=tips, y='day', x='tip', estimator=np.median)
plt.title('Barplot: Median Tip Each Day by Gender', size=15)
plt.show()
```

### Create Barplot using Pandas

```python
tips_by_day = tips[['tip', 'day']].groupby(['day'], as_index=False).mean()
tips_by_day
```

```python
tips_by_day.plot.bar(x='day', y='tip')
plt.title('Barplot: Average Tip Each Day', size=15)
plt.show()
```
