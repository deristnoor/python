# 1. Histogram

# 3. Lineplot

### Create Lineplot using Matplotlib

```python
import matplotlib.pyplot as plt
import seaborn as sns

flights = sns.load_dataset('flights')
flights.head()
```

```python
x = flights[flights['month'] == 'Jan']['year']
y = flights[flights['month'] == 'Jan']['passengers']

plt.figure(figsize=(8,6))
plt.plot(x, y, 'blue')
plt.title('Number of Passengers Trends in January', size=15)
plt.xlabel('Year')
plt.ylabel('Number of Passengers')
plt.show()
```

```python
x1 = flights[flights['month'] == 'Jan']['year']
y1 = flights[flights['month'] == 'Jan']['passengers']
x2 = flights[flights['month'] == 'Dec']['year']
y2 = flights[flights['month'] == 'Dec']['passengers']

plt.style.use('seaborn')
plt.figure(figsize=(8,6))
plt.plot(x1, y1, 'blue', 
         x2, y2, 'red')
plt.title('Number of Passengers Trends in January & December', size=15)
plt.xlabel('Year')
plt.ylabel('Number of Passengers')
plt.legend(['Januar Passengers', 'December Passengers'])
plt.show()
```

### Create Lineplot using Seaborn

```python
sns.lineplot(data=flights, x='year', y='passengers')
plt.title('Number of Passengers Trends', size=15)
plt.show()
```

```python
sns.lineplot(data=flights, x='year', y='passengers', hue='month')
plt.title('Number of Passengers Trends by Months', size=15)
plt.show()
```

### Create Lineplot using Pandas

```python
df_jan = flights[flights['month'] == 'Jan']
df_jan.plot.line(x='year', y='passengers')
plt.title('Number of Passengers Trends in January', size=15)
plt.show()
```

# 5. Bar Plot

### Create Barplot using Matplotlib

```python
import matplotlib.pyplot as plt
import seaborn as sns

tips = sns.load_dataset('tips')
tips.head()
```

```python
tips_by_day = tips[['tip', 'day']].groupby(['day']).mean()
tips_by_day
```

```python
x = tips_by_day.index.tolist()
y = tips_by_day['tip']

plt.figure(figsize=(8,6))
plt.bar(x, y, color=['green', 'blue', 'red', 'orange'])
plt.title('Barplot: Average Tip Each Day', size=15)
plt.xlabel('Day')
plt.ylabel('Tip')
plt.show()
```

```python
x = tips_by_day.index.tolist()
y = tips_by_day['tip']

plt.figure(figsize=(8,6))
plt.barh(x, y, color=['green', 'blue', 'red', 'orange'])
plt.title('Barplot: Average Tip Each Day', size=15)
plt.xlabel('Tip')
plt.ylabel('Day')
plt.show()
```

```python
x = tips_by_day.index.tolist()
y = tips_by_day['tip']

plt.style.use('seaborn')
plt.figure(figsize=(8,6))
plt.bar(x, y, color=['green', 'blue', 'red', 'orange'])
plt.title('Barplot: Average Tip Each Day', size=15)
plt.xlabel('Day')
plt.ylabel('Tip')
plt.show()
```

### Create Barplot using Seaborn

```python
sns.barplot(data=tips, x='day', y='tip')
plt.title('Barplot: Average Tip Each Day', size=15)
plt.show()
```

```python
sns.barplot(data=tips, x='day', y='tip', hue='sex')
plt.title('Barplot: Average Tip Each Day by Gender', size=15)
plt.legend(loc=0)
plt.show()
```

```python
sns.barplot(data=tips, y='day', x='tip', hue='sex')
plt.title('Barplot: Average Tip Each Day by Gender', size=15)
plt.legend(loc=0)
plt.show()
```

```python
import numpy as np

sns.barplot(data=tips, y='day', x='tip', estimator=np.median)
plt.title('Barplot: Median Tip Each Day by Gender', size=15)
plt.show()
```

### Create Barplot using Pandas

```python
tips_by_day = tips[['tip', 'day']].groupby(['day'], as_index=False).mean()
tips_by_day
```

```python
tips_by_day.plot.bar(x='day', y='tip')
plt.title('Barplot: Average Tip Each Day', size=15)
plt.show()
```

# 6. Pie Chart

### Create Pie Chart using Matplotlib

```python
import matplotlib.pyplot as plt
import seaborn as sns

tips = sns.load_dataset('tips')
tips.head()
```

```python
tips_by_day = tips[['tip', 'day']].groupby(['day'], as_index=False).mean()
tips_by_day
```

```python
y = tips_by_day['day']
x = tips_by_day['tip']

plt.figure(figsize=(8,6))
plt.pie(x, labels=y)
plt.title('Pie Chart: Average Tip Each Day', size=15)
plt.axis('equal')
plt.show()
```

```python
y = tips_by_day['day']
x = tips_by_day['tip']

plt.figure(figsize=(8,6))
plt.pie(x, labels=y, autopct ='%.2f', explode = (0,0,0,0.1))
plt.title('Pie Chart: Average Tip Each Day', size=15)
plt.axis('equal')
plt.show()
```

### Create Pie Chart using Pandas

```python
plt.style.use('seaborn')
tips_by_day.plot.pie(y='tip', figsize=(8,6), autopct='%.2f', explode=(0,0,0,0.1))
plt.axis('equal')
plt.title('Barplot: Average Tip Each Day', size=15)
plt.show()
```

# 7. Heatmap

### Create Heatmap using Matplotlib

```python
import matplotlib.pyplot as plt
import seaborn as sns

tips = sns.load_dataset('tips')
tips.head()
```

```python
corr_matrix = tips[['total_bill', 'tip', 'size']].corr()
corr_matrix
```

```python
plt.figure(figsize=(8,6))
plt.pcolor(corr_matrix, cmap='RdYlGn')
plt.title('Correlation Map: Total Bill, Tip, Meal Size', size=15)
plt.show()
```

### Create Heatmap using Seaborn

```python
plt.figure(figsize=(8,6))
sns.heatmap(corr_matrix, annot=True, cmap='bone_r')
plt.title('Correlation Map: Total Bill, Tip, Meal Size', size=15)
plt.show()
```

### Choropleth Map

Data Source:
- Jakarta Open data: https://data.jakarta.go.id/dataset/data-jumlah-pelaporan-kedatangan-penduduk-dari-luar-provinsi-dki-jakarta-per-kelurahan-tahun-2021
- Jakarta by Kelurahan geo public: https://pstyd.carto.com/tables/jakarta_by_kelurahan_geo/public

```python
df = pd.read_csv('./datasets/data-jumlah-pelaporan-kedatangan-penduduk-dari-luar-provinsi-dki-jakarta-per-kelurahan-bulan-januari-tahun-2021.csv')
df.head()
```

```python
df_kel = df[['kelurahan', 'jumlah']]
df_kel
```

```python
# jakarta geojson
jakarta = r'./datasets/jakarta_by_kelurahan_geo.geojson'
jkt_location = [-6.2088, 106.8456]

jkt_map = folium.Map(location=jkt_location, zoom_start=11)

folium.Choropleth(geo_data = jakarta,
                 data=df_kel,
                 columns=['kelurahan', 'jumlah'],
                 key_on='feature.properties.name',
                 fill_color='YlOrRd',
                 fill_opacity=0.7,
                 line_opacity=0.2,
                 legend_name='Urban Rate in jakarta'
                 ).add_to(jkt_map)

jkt_map
```

```python
# jakarta geojson

bins = list(df_kel['jumlah'].quantile([0, 0.25, 0.5, 0.75, 1]))

jakarta = r'./datasets/jakarta_by_kelurahan_geo.geojson'
jkt_location = [-6.2088, 106.8456]

jkt_map = folium.Map(location=jkt_location, zoom_start=11)

folium.Choropleth(geo_data = jakarta,
                 data=df_kel,
                 columns=['kelurahan', 'jumlah'],
                 key_on='feature.properties.name',
                 fill_color='BuPu',
                 fill_opacity=0.7,
                 line_opacity=0.2,
                 legend_name='Urban Rate in jakarta',
                 bins=bins,
                 reset=True
                 ).add_to(jkt_map)

jkt_map
```
