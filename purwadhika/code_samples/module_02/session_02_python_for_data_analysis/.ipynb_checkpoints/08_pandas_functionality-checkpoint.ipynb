## Missing Value

```python
df = pd.DataFrame(
    {
        'A':[1,2,np.nan],
        'B':[5,np.nan,np.nan],
        'C':[1,2,3],
    }
)

df
```

```python
df.isna()
```

```python
df.isna().sum()
```

```python
df.dropna()
```

```python
df.dropna(axis=1)
```

```python
df.dropna(thresh=2)
```

```python
df.fillna(value='FILL VALUE')
```

```python
df['A'].fillna(value=df['A'].mean())
```

## Aggregating

```python
data = {
    'Company':['GOOG', 'GOOG', 'MSFT', 'MSFT', 'FB', 'FB'],
    'Person':['Sam', 'Charlie', 'Amy', 'Vanessa', 'Carl', 'Sarah'],
    'Sales':[200,120,340,123,243,350]
}

df_sales = pd.DataFrame(data)
df_sales
```

```python
df_sales.groupby('Company')
```

```python
by_comp = df_sales.groupby('Company')
by_comp.mean()
```

```python
by_comp.std()
```

```python
by_comp.min()
```

```python
by_comp.max()
```

```python
by_comp.count()
```

```python
by_comp.describe()
```

```python
by_comp.describe().transpose()
```

```python
by_comp.describe().transpose()['GOOG']
```
