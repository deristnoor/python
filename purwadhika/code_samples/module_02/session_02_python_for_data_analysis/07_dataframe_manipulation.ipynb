# DataFrame Manipulation

```python
import numpy as np
import pandas as pd
```

```python
from numpy.random import randn
np.random.seed(101)
```

```python
df = pd.DataFrame(
    randn(5,4),
    index='A B C D E'.split(),
    columns='W X Y Z'.split()
)

df
```

## Adding data

```python
df.loc['new'] = [1,2,3,4]
df
```

```python
df['new'] = df['W'] + df['Y']
df
```

```python
df.insert(2, 'new', [1,2,3,4,5])
df
```

## Delete data

```python
df.drop('new', axis=1)
```

```python
df
```

```python
df.drop('new', axis=1, inplace=True)
```

```python
df
```

```python
df.drop('E', axis=0, inplace=True)
df
```

## Multi Indexing

```python
outside = ['Jakarta', 'Jakarta', 'Jakarta', 'Surabaya', 'Surabaya', 'Surabaya']
inside = [1,2,3,1,2,3]
hier_index = list(zip(outside,inside))
hier_index
```

```python
hier_index = pd.MultiIndex.from_tuples(hier_index)
hier_index
```

```python
df = pd.DataFrame(np.random.randint(
    1,100,(6,2)),
    index=hier_index,
    columns=['Restoran A', 'Restoran B']
)
```

```python
df
```

```python
df.loc['Jakarta']
```

```python
df.loc['Jakarta'].iloc[1]
```

```python
df.xs('Jakarta')
```

```python
df.xs(('Jakarta', 1))
```

```python
df.index.names=['City', 'Location']
df
```

```python
df.xs(1, level='Location')
```
