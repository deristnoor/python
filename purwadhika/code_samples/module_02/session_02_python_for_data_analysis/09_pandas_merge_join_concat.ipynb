# Merge, Join, & Concatenate

```python
import pandas as pd
import numpy as np
```

## Merge

```python
left = pd.DataFrame(
    {
        'key':['k0','k1','k2','k3'],
        'A':['a0','a1','a2','a3'],
        'B':['b0','b1','b2','b3']
    }
)
```

```python
left
```

```python
right = pd.DataFrame(
    {
        'key':['k0','k1','k3','k4'],
        'C':['c0','c1','c2','c3'],
        'D':['d0','d1','d2','d3']
    }
)
```

```python
right
```

```python
pd.merge(left, right, on='key')
```

```python
pd.merge(left, right, how='outer', on='key')
```

```python
pd.merge(left, right, how='left', on='key')
```

```python
pd.merge(left, right, how='right', on='key')
```

```python
left
```

```python
right
```

## Join

```python
left1 = pd.DataFrame(
    {
        'A':['a0','a1','a2','a3'],
        'B':['b0','b1','b2','b3']
    }, index=['k0','k1','k2','k3']
)
```

```python
right1 = pd.DataFrame(
    {
        'C':['c0','c1','c2','c3'],
        'D':['d0','d1','d2','d3']
    }, index=['k0','k1','k3','k4'],
)
```

```python
left1.join(right1)
```

```python
right1.join(left1)
```

```python
left1.join(right1, how='outer')
```

## Concatenate

```python
left
```

```python
right
```

```python
pd.concat([left,right])
```

```python
pd.concat([left,right], axis=1)
```
