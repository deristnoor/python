### Data Preparation

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import folium

pd.set_option('display.max_columns', 500)
```

```python
df = pd.read_csv('./datasets/melb_data.csv')
df.head()
```

#### Exercise 1
- Buatlah histogram untuk menampilkan distribusi data harga rumah (kolom <b>Price</b>)
- Tambahkan judul "Distribusi Harga Rumah" berukuran 15
- Tidak ada ketentuan pilihan warna, ukuran dan library yang digunakan

```python
plt.figure(figsize=(8,6))
sns.histplot(df['Price'])
plt.title('Distribusi Harga Rumah', size=15)
plt.show()
```

#### Exercise 2
- Buatlah box plot untuk menampilkan distribusi data jarak ke CBD (kolom <i><b>Distance</b></i>)
- Buatlah box plot <i><b>Distance</b></i> untuk tiap nama region (kolom <i><b>Regionnname</b></i>)
- Tambahkan judul "Distribusi Jarak ke CBD berdasarkan Nama Region" berukuran 15
- Tidak ada ketentuan pilihan warna, ukuran dan library yang digunakan

```python
plt.figure(figsize=(8,6))
sns.boxplot(x=df['Distance'], y=df['Regionname'])
plt.title('Distribusi Jarak ke CBD berdasarkan Nama Region', size=15)
plt.show()
```

#### Exercise 3
- Buatlah bar plot untuk membandingkan nilai median harga rumah (kolom <i><b>Price</b></i>) dan tiap tipe rumah (kolom <i><b>Type</b></i>)
- Tambahkan judul "Perbandingan Nilai Median Harga Rumah" berukuran 15
- Tidak ada ketentuan pilihan warna, ukuran dan library yang digunakan

```python
plt.figure(figsize=(8,6))
sns.barplot(data=df, x='Type', y='Price', estimator=np.median)
plt.title('Perbandingan Nilai Median Harga Rumah', size=15)
plt.show()
```

#### Exercise 4
- Buatlah scatter plot untuk menampilkan korelasi antara harga rumah (kolom <i><b>Price</b></i>) dengan jarak ke CBD (kolom <i><b>Distance</b></i>)
- Bedakan warna titik berdasarkan tipe rumah (kolom <i><b>Type</b></i>)
- Tambahkan judul "Korelasi Harga Rumah dan Jarak Rumah ke CBD" berukuran 15
- Tidak ada ketentuan pilihan warna, ukuran dan library yang digunakan

```python
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x='Price', y='Distance', hue='Type')
plt.title('Korelasi Harga Rumah dan Jarak Rumah ke CBD', size=15)
plt.show()
```

#### Exercise 5
- Konversi kolom <i><b>Date</b></i> menjadi tipe data datetime64. Buatlah kolom baru bernama 'Month' hasil ekstraksi data bulan dari kolom <i><b>Date</b></i>
- Buatlah line plot untuk menampilkan tren perkembangan harga rumah (kolom <i><b>Price</b></i>) tiap bulan 
- Bedakan warna garis berdasarkan tipe rumah (kolom <i><b>Type</b></i>)
- Tambahkan judul "Tren Harga Rumah per Bulan berdasarkan Tipe Rumah" berukuran 15
- Tidak ada ketentuan pilihan warna, ukuran dan library yang digunakan

```python
from datetime import datetime as dt
df['Month'] = pd.to_datetime(df['Date']).dt.strftime('%m')
df.head()
```

```python
plt.figure(figsize=(8,6))
sns.lineplot(data=df, x='Month', y='Price', hue='Type')
plt.title('Tren Harga Rumah per Bulan berdasarkan Tipe Rumah', size=15)
plt.show()
```

#### Exercise 6
- Kelompokkan total luas tanah (kolom <i><b>Landsize</b></i>) berdasarkan region (kolom <i><b>Regionnname</b></i>), lalu simpan ke dalam sebuah DataFrame baru 
- Dari DataFrame tersebut, buatlah pie chart unutk menampilkan proporsi total luas tanah di tiap region
- Tambahkan explode agar potongan diagram seakan terpisah
- Tambahkan judul "Proporsi Luas Tanah per Region" berukuran 20
- Tidak ada ketentuan pilihan warna, ukuran dan library yang digunakan

```python
landsum_per_region = df.groupby(['Regionname'])['Landsize'].sum().reset_index()

y = landsum_per_region['Regionname']
x = landsum_per_region['Landsize']

plt.figure(figsize=(8,6))
plt.pie(x, labels=y, autopct='%.2f', explode=(0.05,0.05,0.05,0.05,0.05,0.05,0.05,0.05,))
plt.title('Proporsi Luas Tanah per Region', size=15)
plt.axis('equal')
plt.show()
```

#### Exercise 7
- Buatlah terlebih dahulu <i><b>correlation matrix</b></i> yang berisi korelasi kolom <i><b>Price</b></i>, <i><b>Distance</b></i>, <i><b>Landsize</b></i>, dan <i><b>Rooms</b></i>
- Dari <i><b>correlation matrix</b></i> ini, buatlah <i><b>correlation heatmap</b></i> unutk menggambarkan korelasi antar variabel
- Tampilkan data korelasi di <i><b>heatmap</b></i> (annot = True)
- Tambahkan judul "Heatmap Korelasi Harga, Jarak, Luas Tanah dan Jumlah Kamar" berukuran 20
- Tidak ada ketentuan pilihan warna, ukuran dan library yang digunakan

```python
corr_matrix = df[['Price', 'Distance', 'Landsize', 'Rooms']].corr(method='pearson')

plt.figure(figsize=(12,10))
sns.heatmap(corr_matrix, annot=True, cmap='twilight_shifted')
plt.title('Heatmap Korelasi Harga, Jarak, Luas Tanah dan Jumlah Kamar', size=20)
plt.show()
```

#### Exercise 8
- Urutkan data berdasarkan harga (kolom <i><b>Price</b></i>) dan tampilkan 10 data dari harga rumah tertinggi, kemudian simpan hasilnya ke dalam DataFrame baru 
- Buatlah peta Melbourne yang berlokasi di [-37.8136, 144.9631] dan zoom_start = 11
- Didalam peta Melbourne, tambahkan 10 marker sesuai lokasi 10 rumah dengan harga tertinggi
- didalam marker, popup berisi harga rumah dan tooltip berisi alamat rumah

```python
top_10_price = df.sort_values(['Price'], ascending=False).head(10).reset_index()
top_10_price
```

```python
# melbourne map
melbourne_map = folium.Map(location=[-37.8136, 144.9631], zoom_start=11)

# marker
for i in range(len(top_10_price['Price'])):
               folium.Marker([top_10_price['Lattitude'][i], top_10_price['Longtitude'][i]], 
                             popup='Price: {}'.format(top_10_price['Price'][i]), 
                             tooltip='Address: {}'.format(top_10_price['Address'][i])
                            ).add_to(melbourne_map)
melbourne_map
```
